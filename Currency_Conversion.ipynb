```python
from langchain_google_genai import ChatGoogleGenerativeAI

llm = ChatGoogleGenerativeAI(
    model="gemini-2.5-flash",   # Changed model to gemini-2.5-flash
    temperature=0.2
)
```

```python
from langchain_core.tools import tool
import requests

@tool
def get_conversion_factor(base_currency: str, target_currency: str) -> float:
    """
    Fetch currency conversion factor between base and target currency.
    """
    url = f'https://v6.exchangerate-api.com/v6/c754eab14ffab33112e380ca/pair/{base_currency}/{target_currency}'
    response = requests.get(url).json()
    # Return the conversion_rate directly, as intended
    return response['conversion_rate']

@tool
def convert(base_currency_value: float, conversion_rate: float) -> float:
    """
    Convert given base currency value into target using conversion rate.
    """
    return base_currency_value * conversion_rate
```

```python
from langchain.agents import initialize_agent, AgentType

agent = initialize_agent(
    tools=[get_conversion_factor, convert],
    llm=llm,
    agent=AgentType.STRUCTURED_CHAT_ZERO_SHOT_REACT_DESCRIPTION,
    verbose=True
)
```

```python
response = agent.invoke({
    "input": "What is the conversion factor between INR and USD, and convert 10 INR to USD?"
})

print(response["output"])
```

```python
from langchain_community.tools import DuckDuckGoSearchRun
from langchain_community.utilities import DuckDuckGoSearchAPIWrapper

search_wrapper = DuckDuckGoSearchAPIWrapper()
search_tool = DuckDuckGoSearchRun(api_wrapper=search_wrapper)
```

```python
@tool
def get_weather_data(city: str) -> str:
  """
  This function fetches the current weather data for a given city
  """
  url = f'https://api.weatherstack.com/current?access_key=4d1d8ae207a8c845a52df8a67bf3623e&query={city}'

  response = requests.get(url)

  return response.json()
```

```python
from langchain.agents import create_react_agent, AgentExecutor
from langchain import hub
```

```python
!pip install langchainhub
# Step 2: Pull the ReAct prompt from LangChain Hub
prompt = hub.pull("hwchase17/react")  # pulls the standard ReAct agent prompt
```

```python
# Step 3: Create the ReAct agent manually with the pulled prompt
agent = create_react_agent(
    llm=agent,
    tools=[search_tool, get_weather_data],
    prompt=prompt
)
```

```python
# Step 4: Wrap it with AgentExecutor
agent_executor = AgentExecutor(
    agent=agent,
    tools=[search_tool, get_weather_data],
    verbose=True
)
```

```python
# Step 5: Invoke
response = agent_executor.invoke({"input": "Find the capital of Madhya Pradesh, then find it's current weather condition"})
print(response)
```
